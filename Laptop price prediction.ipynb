```python
from scipy.stats import skew

#finding the skewness in each variable
data_num_skew = X.apply(lambda i: skew(i.dropna())) #dropping the "na" values from the data and then passing it to the skew function to calculate skewness of the variable

#subsetting the variables that are highly skewed  # Filtering records from more no. of variables
data_num_skewed = data_num_skew[(data_num_skew > .75) | (data_num_skew < -.75)]
 
print(data_num_skew)
print()
print(data_num_skewed)

import numpy as np

# apply log + 1 transformation for all numeric features with skewness over .75
X[data_num_skewed.index] = np.log1p(X[data_num_skewed.index])   
```

### Splitting data into Train and test

```python
from sklearn.model_selection import train_test_split

# Split the data into Train and Test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
# print("Percent of train data", X_train.shape[0]/X.shape[0]*100)
```

```python
from sklearn.linear_model import LinearRegression

# Create a model object
lm = LinearRegression()

# train the model object
lm.fit(X_train, Y_train)

# print intercept and coefficients
print(lm.intercept_)      # bestfit line intersecting on the Y-axis (B0 value)
print(lm.coef_)           # B1, B2, B3 values 
```

```python
# pair the feature name swith the coefficients
print(list(zip(X.columns, lm.coef_)))
```

```python
# predict using the model
Y_pred = lm.predict(X_test)
print(Y_pred)
```

```python
new_df=pd.DataFrame()
new_df=X_test.copy()
 
new_df["Actual sales"]=Y_test
new_df["Predicted sales"]=Y_pred
new_df=new_df.reset_index().drop("index", axis=1)
```
